```python
import pandas as pd
```

```python
import numpy as np

# Create fake dataframe with object image data
df = pd.DataFrame({
    'image_id': range(1, 11),
    'filename': [f'image_{i}.jpg' for i in range(1, 11)],
    'object_category': ['dog', 'dog', 'cat', 'cat', 'bird', 'dog', 'car', 'car', 'bicycle', 'bird'],
    'familiarity': ['familiar', 'familiar', 'familiar', 'novel', 'novel', 'familiar', 'familiar', 'novel', 'novel', 'familiar'],
    'image_array': [np.random.rand(224, 224, 3) for _ in range(10)],
    'confidence_score': np.random.uniform(0.5, 0.99, 10)
})

print(df)
```

```python
#extract just familiar
familiar_df = df[df['familiarity'] == 'familiar']
print("\nFamiliar Images:\n", familiar_df)

#grab random single image
random_image = df.sample(n=1)
print("\nRandom Image:\n", random_image)
```

```python
# novel in the same category but different image
novel_df = df[(df['familiarity'] == 'novel') & (df['object_category'] == 'cat')]
print("\nNovel Images in 'cat' category:\n", novel_df)
```

```python
# make a list of 24 5s and 24 6s
fives_and_sixes = [5]*24 + [6]*24

# make a list of 12 1s, 12 2s, 12 3s, and 12 4s
ones_to_fours = [1]*12 + [2]*12 + [3]*12

#combine them
combined_list = fives_and_sixes + ones_to_fours
```

```python
#shuffle the combined list
np.random.shuffle(combined_list)
```

```python
#create a loop where every 5th item shows somethign from combined_list
n = 0
for i in range(0, 150):
    if (i+1) % 5 == 0:
        print(combined_list[n])
        n += 1
    else:
        print("other img")
```
